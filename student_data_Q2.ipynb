# ---------------------------------------------------------------------------------------------------------------

 Where there was a missing value in the data set filled the missing value with zero.

```python
new_Data = data.fillna(0)
```

```python
new_Data.isnull().sum()
```

```python
new_Data2 = new_Data[['Student Name','Total (100)','Grade']]
new_Data2
```

```python
grade_coun = new_Data2["Grade"].value_counts()
print(grade_coun)
```

```python
new_Data2.rename(columns={'Total (100)': 'Total_Marks'},inplace=True)
```

```python
grade_graph = grade_coun.plot.bar(x = 'Grade', y = 'Total_Marks', fontsize='12')
```

```python
GradeA = new_Data2[new_Data2.Grade == "A"]
```

```python
GradeA.sort_values(by='Total_Marks',ascending=False)
```

```python
GradeB = new_Data2[new_Data2.Grade == "B"]
```

```python
GradeB.sort_values(by="Total_Marks",ascending=False)
```

```python
GradeC = new_Data2[new_Data2.Grade == "C"]
GradeC.sort_values(by="Total_Marks",ascending=False)
```

```python
Fail = new_Data2[new_Data2.Grade == "F"]
Fail.sort_values(by="Total_Marks",ascending=False)
```
